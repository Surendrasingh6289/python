```python
import pandas as pd
import numpy as np
import warnings
```

```python
df = pd.read_csv(r'D:\Grow Tech\Prcatice_data_cleaning\Uncleaned Dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
warnings.filterwarnings('ignore')
```

```python
df['Units_Sold']=df['Units_Sold'].fillna(df['Units_Sold'].mean(),inplace=True)
df['Unit_Price']=df['Unit_Price'].fillna(df['Unit_Price'].mean(),inplace=True)
df['Discount_%']=df['Discount_%'].fillna(df['Discount_%'].mean(),inplace=True)
df['Order_Date']=df['Order_Date'].fillna(df['Order_Date'].mode(),inplace=True)
df['Segment']=df['Segment'].fillna(df['Segment'].mode(),inplace=True)
```
