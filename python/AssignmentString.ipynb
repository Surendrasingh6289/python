```python
#1️1) Question: Count uppercase and lowercase letters.
n = input()
print(n)
count_upper = 0
count_lower = 0
for i in n:
    if i.isupper()==True:
        count_upper+=1
    if i.islower()==True:
        count_lower+=1
print(count_upper)
print(count_lower)
```

```python
#3️1️) Question: Convert string to lowercase without lower().

n = input()
l = ''
for i in n:
    if ord(i)>=65 and ord(i)<=91:
        l += chr(ord(i)+32)
    else:
        l+=i
print(l)
```
