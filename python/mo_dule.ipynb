**MATH MODULE**

```python
import math
```

```python
x = [56,34,6,97]
print(math.fsum(x))
print(int(math.fsum(x)))
```

```python
from math import fsum
print(fsum([56,34,7,65]))
```

```python
x = float(input())
print(round(x))
print(math.floor(x))
print(math.ceil(x))
print(abs(x))
print(math.fabs(x))
```

```python
print(math.factorial(5))
print(math.pow(2,4))
print(math.sqrt(9))
print(math.sqrt(34.53))
print(math.cbrt(8))
```

**DATETIME MODULE**

```python
from datetime import datetime
```

```python
x = datetime.today()
print(x.ctime())
print(x.date())
print(x.day)
print(x.month)
print(x.year)
print(x.hour)
print(x.minute)
print(x.second)
print(x.microsecond)
```

```python
print(x.strftime('%A'))
print(x.strftime('%a'))
print(x.strftime('%B'))
print(x.strftime('%b'))
print(x.strftime('%C'))
print(x.strftime('%c'))
print(x.strftime('%D'))
print(x.strftime('%d'))
print(x.strftime('%M'))
print(x.strftime('%m'))
print(x.strftime('%x'))
print(x.strftime('%X'))
print(x.strftime('%Y'))
print(x.strftime('%y'))
print(x.strftime('%H'))
print(x.strftime('%h'))
print(x.strftime('%S'))
print(x.strftime('%T'))
```
