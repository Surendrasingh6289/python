```python

s = input()
vowel_count=0
consonant_count = 0
for i in s.lower():
    if i=='a'or i=='e'or i=='i'or i=='o'or i=='u':
        vowel_count+=1
    else:
        consonant_count+=1
print(vowel_count)
print(consonant_count)
```
