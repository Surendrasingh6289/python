```python
def Gender_encode(x):
    if x =='Male':
        return 'M'
    else:
        return 'F'
```

```python
mall['Gender_category']= mall['Gender'].apply(Gender_encode)
mall.head()
```

```python
def Income_cat(x):
    if x > 0 and x < 30:
        return "Lower class"
    elif x>=30 and x<75:
        return 'Middle class'
    else:
        return 'High class'
```

```python
mall['Income Category'] = mall['Annual Income'].apply(Income_cat)
mall
```

```python
concat_df = pd.concat((iris,mall),axis=1)
concat_df
```
