```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
peng = pd.read_csv(r'D:\Grow Tech\python\new_penguin.csv',index_col='Unnamed: 0')
```

```python
peng.info()
```

```python
plt.plot(peng['bill_length_mm'],marker = '.',mec = 'red',ls = 'dotted')
plt.show()
```

```python
plt.figure(figsize=(6,2))
plt.plot(peng['bill_length_mm'])
plt.show()
```

```python
plt.figure(figsize=(6,2))
plt.plot(peng['bill_length_mm'],label = "bill_length")
plt.plot(peng['flipper_length_mm'],label = "flipper_length")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,2))
plt.plot(peng['bill_length_mm'],label = "bill_length",color ='red')
plt.plot(peng['flipper_length_mm'],label = "flipper_length",color ='green')
plt.title('line chart')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar(x=peng['island'],height=peng['flipper_length_mm'])
plt.xlabel('island')
plt.ylabel('flipper_length')
plt.show()
```
