```python
x = int(input("Enter value: "))

if x > 0 :
    print("Positive",x)
elif x==0:
    print("Zero",x)
else :
    print("Negative",x)
```

```python
x = int(input("Enter marks: "))

if x >= 75 :
    print("First division")
elif x >= 50 and x < 75:
    print("Second division")
elif x >= 35 and x < 50:
    print("Third division")
elif x<0 or x>100:
    print("Invalid Marks")
else:
    print("Failed")
```

```python
#even odd

x = int(input("Enter values: "))

if x % 2 ==0:
    print("Even")
else:
    print("Odd")
```

```python
#leap Year

x = int(input("Enter year: "))

if (x % 4==0 and x % 100 !=0) or (x % 400 == 0):
    print("Leap year",x)
else:
    print("Not leap year",x)
```

```python
sales = int(input("Enter sales: "))
salary = 10000

if sales >= 50000:
    salary += int(salary*(10/100))
    print("Increased salary is ",salary)
else:
    print("No increment")
```

```python
a = int(input("Enter First angel: "))
b = int(input("Enter second angel: "))
c = int(input("Enter third angel: "))

if a + b + c == 180:
    print("Is Triangle")
else:
    print("Not a Triangle")
```

```python
n = int(input())
if n >= 0:
    if n ==0:
        print("zero")
    else:
        print('positive')
else:
    print("negative")
```

```python
age = int(input("Enter age: "))
eligibility = True

if eligibility:
    if age >0:
        if age >= 18 and age<=100:    
            print("Eligible  for license")
        else:
            print("under age")
    else:
        print("Enter valid input")
else:
    print("not eligible")
```

```python
balance = int(input('balance'))
withdraw_amount = int(input('withdraw amount'))
daily_limit = int(input('daily limit'))

if withdraw_amount< balance:
    if withdraw_amount <= daily_limit:
        print('amount withdrawn')
    else:
        print('daily limit exceeded')
else:
    print('insufficient balance')
```
