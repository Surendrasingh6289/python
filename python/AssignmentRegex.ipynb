```python
import re
import warnings
warnings.filterwarnings("ignore")
```

```python
#Question 19: Problem: Replace all occurrences of "color" with "colour", but only if "color" is a whole word. 
l =  "The color is red. This coloring book has many colors."
pattern = r'\bcolor\b'
print(re.sub(pattern,'colour',l))
```

```python
#Question 18: Problem: Find words that are exactly 5 characters long. 
l = "Apple, Banana, Grape, Cherry, Mango"
pattern = r'\b[A-Za-z]{5}\b' # r'\b\w....\b'
print(re.findall(pattern,l))
```

```python
#Question 20: Problem: Extract text between <start> and <end> tags. 
l = "This is <start>important data</end> and this is <start>more data</end>."
pattern = '<start>(.*?)</end>'
pattern1 = '</\w+>'
print(re.findall(pattern,l))
```

```python
#Question 22: Problem: Extract all numbers that are between 1 and 100 (inclusive). 
l = "Numbers are 5, 105, 99, 0, 750, and 20."
pattern = r'\b(100|[1-9]\d?)\b'
print(re.findall(pattern,l))
```

```python
#Question 23: Problem: Find all words that have exactly two vowels.
l = "Beautiful, apple, tree, sky, rhythm, fantastic"
pattern = r'\b[b-df-hj-np-tv-z]*[aeiou][b-df-hj-np-tv-z]*[aeiou][b-df-hj-np-tv-z]*\b'
print(re.findall(pattern,l))
```

```python
#Question 24: Problem: Validate if a string represents a time in 24-hour format (HH:MM). Test
S1= "14:30" 
S2= "25:00"
pattern = r'([01][0-9]|2[0-3]:[0-5][0-9])'    # r'([0-2][0-9]:[0-5][0-9])'
print(re.match(pattern,S1))
print(re.match(pattern,S2))
```

```python
#Question 25: Problem: Extract all hexadecimal color codes (e.g., #RRGGBB or #RGB). Test
l =  "Colors include #FF0000, #00F, and #AABBCC."
pattern = r'(#[0-9A-Fa-f]{3}|#[0-9A-Fa-f]{6})\b'
print(re.findall(pattern,l))
```

```python
# Question 26: Problem: Find all words that start and end with the same letter. Test String:
l = "Level, Madam, Racecar, Noon, Hello, World"
pattern = r'(\w[a-zA-Z]+\w)'
print(re.findall(pattern,l))
```
