```python
#1. Create a dictionary from two lists
keys = ['a','b','c']
values = [1,2,3]
di = {}
for i in range(len(values)):
    di[keys[i]] = values[i]
print(di)
```

```python
#2. Count frequency of each character in a string
s = "banana"
freq = {}

for i in s:
    if i in freq:
        freq[i]+=1
    else:
        freq[i]=1
print(freq)
```

```python
#3. Merge two dictionaries
l = {'x':1, 'y':2}
m = {'z':3}
l.update(m)
print(l)
```

```python
#4. Remove a key from a dictionary
l = {'a':1, 'b':2}
l.pop('a')
print(l)
```

```python
#5. Find the key with the maximum value 
l = {'a':10, 'b':30, 'c':20} 
maxi_value = -1
maxi_keys = None
for k in l:
    if l[k] > maxi_value :
        maxi_value = l[k]
        maxi_keys = k

print(maxi_keys)
```

```python
#6. Find the sum of all dictionary values 
l = {'a':1, 'b':2, 'c':3} 
sums = 0
for i in l.values():
    sums+=i
print(sums)
```

```python
#11. Find average of dictionary values 
l = {'a':10, 'b':20, 'c':30} 
sums = 0
avg = 0
count = 0
for value in l.values():
    sums += value
    count +=1
avg = sums/count
print(avg)
```

```python
#12. Sort a dictionary by its values 
l = {'a':3, 'b':1, 'c':2} 
d_item = list(l.items())
for i in range(len(d_item)):
    for j in range(i+1,len(d_item)):
        if d_item[i][1]>d_item[j][1]:
            d_item[i],d_item[j]=d_item[j],d_item[i]
print(dict(d_item))
    
```

```python
#13. Sort a dictionary by its keys 
l = {'b':2, 'a':1} 
d_item = list(l.items())
for i in range(len(d_item)):
    for j in range(i+1,len(d_item)):
        if d_item[i][0]>d_item[j][0]:
            d_item[i],d_item[j]=d_item[j],d_item[i]
print(dict(d_item))
```

```python
#14. Convert a formatted string to a dictionary 
s ='a=1, b=2' 
l = s.split()
key = []
value = []
di1 = {}
for i in l:
    for j in i:
        if j.isalpha()==True:
            key.append(j)
        if j.isnumeric()==True:
            value.append(j)
for i in range(len(key)):
    di1[key[i]]=int(value[i])
print(di1)

'''
s ='a=1, b=2' 
di = {}

for item in s.split(','):
    k,v = item.split('=')
    di[k.strip()]=int(v)
print(di)
'''
```
