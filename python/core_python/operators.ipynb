**IDENTITY OPERATOR**

is 

not is


```python
a = 'python'
b = 'python'

print(id(a))
print(id(b))

print(a is b)
print(a is not b)
```

```python
u =[1,2,3]
v =[1,2,3]

print(id(u))
print(id(v))
print(u is v) # checks if reference values are same
print(u is not v)
print(u == v) #checks if values are same
```
