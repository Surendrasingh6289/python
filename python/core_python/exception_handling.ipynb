```python
def extendAP(arr):
    new_arr = []
    dif = arr[1]-arr[0]
    length = len(arr)
    for i in range(length):
        new_arr.append(arr[i])
    for i in range(3):
        new_arr.append(new_arr[-1]+dif)
    return new_arr

array = (1, 5, 9, 13, 17)
extendAP(array)
```
