```python
li =  [1,2,3,4,5,6,11,23,45,67,78,56,34,22]
print(li)
even =[]
odd = []

for i in li:
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)
print(even)
print(odd)
```

```python
li =  [1,2,3,4,5,6,10,7,13,20,100,11,23,45,67,78,56,34,22]
prime = []
for i in li:
    if i>1:
        for j in range(2,i):
            if i % j ==0:
                break
        else:
            prime.append(i)
print(prime)
```
