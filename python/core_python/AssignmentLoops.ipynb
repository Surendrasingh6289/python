```python
#46. Calculate and print the LCM (Least Common Multiple) of two numbers using a while loop. 
x = int(input())
y = int(input())
a,b =x,y
while b!=0:
    a,b = b,a%b
gcd = a
lcm = (x*y)//gcd
print(lcm)
```
