```python
#12. Sort a dictionary by its values 
l = {'a':3, 'b':1, 'c':2} 
d_item = list(l.items())
for i in range(len(d_item)):
    for j in range(i+1,len(d_item)):
        if d_item[i][1]>d_item[j][1]:
            d_item[i],d_item[j]=d_item[j],d_item[i]
print(dict(d_item))
    
```

```python
#13. Sort a dictionary by its keys 
l = {'b':2, 'a':1} 
d_item = list(l.items())
for i in range(len(d_item)):
    for j in range(i+1,len(d_item)):
        if d_item[i][0]>d_item[j][0]:
            d_item[i],d_item[j]=d_item[j],d_item[i]
print(dict(d_item))
```

```python
#14. Convert a formatted string to a dictionary 
s ='a=1, b=2' 
l = s.split()
key = []
value = []
di1 = {}
for i in l:
    for j in i:
        if j.isalpha()==True:
            key.append(j)
        if j.isnumeric()==True:
            value.append(j)
for i in range(len(key)):
    di1[key[i]]=int(value[i])
print(di1)

'''
s ='a=1, b=2' 
di = {}

for item in s.split(','):
    k,v = item.split('=')
    di[k.strip()]=int(v)
print(di)
'''
```

```python
#23. Count number of nested keys in a dictionary  
d = {'a':1, 'b':{'x':2,'y':3}} 
count=0
for i in d:
    if type(d[i])==dict:
        count+=1
        for j in d[i]:
            count+=1
    else:
        count+=1
print(count)

```

```python
def count_keys(d):
    count = 0
    if not isinstance(d,dict):
        return 0
    for values in d.values():
        count+=1
        if isinstance(values,dict):
            count+=count_keys(values)
    return count
d = {'a':1, 'b':{'x':2,'y':3}} 
print(count_keys(d))
```
