```python
#1. Write a Python script to sort (ascending and descending) a dictionary by value.
d1 = {'a':1,'b':2,'c':3}
dict_item = list(d1.items())

for i in range(len(dict_item)):
    for j in range(i+1,len(dict_item)):
        if dict_item[i][1]> dict_item[j][1]:
            dict_item[j],dict_item[i]=dict_item[i],dict_item[j]
print(dict_item)
```

```python
#8. Write a Python program to combine two dictionary by adding values for common keys.
'''d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}
Sample output: Counter({'a': 400, 'b': 400, 'd': 400, 'c': 300})'''
d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}
for i in d2:
    if i in d1:
        d1[i]+=d2[i]
    else:
        d1[i]= d2[i]

print(d1)


```

```python
'''9. Write a Python program to print all distinct values in a dictionary.
Sample Data : [{"V":"S001"}, {"V": "S002"}, {"VI": "S001"}, {"VI": "S005"},
{"VII":"S005"}, {"V":"S009"},{"VIII":"S007"}]
Expected Output : Unique Values: {'S005', 'S002', 'S007', 'S001', 'S009'}'''
d1 = [{"V":"S001"}, {"V": "S002"}, {"VI": "S001"}, {"VI": "S005"},{"VII":"S005"}, {"V":"S009"},{"VIII":"S007"}]
s = set()
for i in d1:
    for key,value in i.items():
        s.add(value)
print(s)
```

```python
'''11. Write a Python program to combine values in a list of dictionaries.
Sample data: [{'item': 'item1', 'amount': 400}, {'item': 'item2', 'amount': 300},
{'item': 'item1', 'amount': 750}]
Expected Output: Counter({'item1': 1150, 'item2': 300})'''

l = [{'item': 'item1', 'amount': 400}, {'item': 'item2', 'amount': 300},{'item': 'item1', 'amount': 750}]
d = {}
for i in l:
    item =i['item']
    amount =i['amount']
    if item in d:
        d[item]+=amount
    else:
        d[item]=amount
print(d)
```

```python
'''12. Write a Python program to create a dictionary from a string.
Note: Track the count of the letters from the string.
Sample string : 'google'
Expected output: {'g': 2, 'o': 2, 'r': 2, 'e': 1, 'l': 1}'''
s = 'google'
d ={}
for i in s:
    if i in d:
        d[i]+=1
    else:
        d[i]= 1
print(d)
```
