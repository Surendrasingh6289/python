```python
#42️. Find difference between sum of even and odd numbers.
l = [1,2,3,4,5]
even = [i for i in l if i%2==0]
odd = [i for i in l if i%2!=0]
even_sum = 0
odd_sum = 0
for i in even:
    even_sum+=i
for i in odd:
    odd_sum+=i
if even_sum>odd_sum:
    print(even_sum-odd_sum)
else:
    print(odd_sum-even_sum)
```
