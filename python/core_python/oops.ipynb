```python
class Grandfather:
    def Dada(self):
        print("dadaji ram ram")
class Father(Grandfather):
    def papa(self):
        print("pitaji pranam")
class Son(Father):
    def beta(self):
        print("wassup")

puttar = Son()
puttar.Dada()
puttar.papa()
puttar.beta()
```

```python
class Category():
    def Cat(self):
        print("this is base class")
class Mouse(Category):
    def Mise(self):
        print("this is Output")
class Marker(Category):
    def Mark(self):
        print("this is marker output")
class Keyboard(Category):
    def Key(self):
        print("this is keyboard output")

kb = Keyboard()
kb.Key()
kb.Cat()
```

```python
class Manager():
    def Man(self):
        print("Manager base class")
class Hr(Manager):
    def h(self):
        print("recruiting department")
class trainer(Manager):
    def t(self):
        print("provides training")
class employee(Hr,trainer):
    def emp(self):
        print('corporate majdoor')

sja = employee()
sja.Man()
sja.h()
```

```python
class Employee:
    def Info(self,name,age):
        self.name = name
        self.age = age
    def give(self):
        print(self.name,self.age)

class Admin():
    def Info(self,name,age):
        self.name = name
        self.age = age
    def give(self):
        print(self.name,self.age)

emp = Employee()
emp.Info('surendra',22)
emp.give()
ceo = Admin()
ceo.Info('virendra',22)
ceo.give()
```

```python
class BMW():
    def model(self,model):
        self.model = model
        print("model is ",model)
    def price(self,price):
        self.price = price
        print("price is ",price)
class Audi():
    def model(self,model):
        self.model = model
        print("model is ",model)
    def price(self,price):
        self.price = price
        print("price is ",price)
class Hyundai():
    def model(self,model):
        self.model = model
        print("model is ",model)
    def price(self,price):
        self.price = price
        print("price is ",price)

car  = BMW()
car.price(13456789)
car.model('BMW X7')

car = Audi()
car.price(112345678)
car.model('Audi Q7')

car = Hyundai()
car.price(1123456)
car.model('Hyundai Verna')
```

```python
class Public():
    def Show(self):
        print("public class")

class Protected():
    def _protect(self):
        print('protected class')

class Private():
    def __pri(self):
        print("private method")

pvt = Private()
pvt._Private__pri()
```

```python
class Public():
    def _Show(self):
        print("public class")
class _Protected(Public):
    def __protect(self):
        print("protected  class")
pub = _Protected()
pub._Show()
pub._Protected__protect()
```
