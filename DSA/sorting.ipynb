**BUBBLE SORT**

```python
arr = [1,23,4,13,11,9,45]
l = len(arr)
for i in range(l-1):
    for j in range(l-i-1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]
print(arr)
```

**SELECTION SORT**

```python
arr = [64, 34, 25, 5, 22, 11, 90, 12]

n = len(arr)
for i in range(n-1):
    min_index= i
    for j in range(i+1,n-1):
        if arr[j]<arr[min_index]:
            min_index = j
    min_value =arr.pop(min_index)
    arr.insert(i,min_value)
print(arr)
```
